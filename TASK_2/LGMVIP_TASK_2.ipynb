##**Importing Essential Libraries**

```python
import cv2
```

```python
from  google.colab.patches import cv2_imshow
```

##**Read The Image**

```python
img=cv2.imread('/content/sample_data/Monkey.jpg')
```

```python
cv2_imshow(img)
```

##**Converting image into Gray Scale**

```python
gray_scale_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale_img)
```

##**Inverting image**

```python
inverted_img=cv2.bitwise_not(gray_scale_img)
cv2_imshow(inverted_img)
```

##**Bluring the Image**

```python
img_smoothing=cv2.GaussianBlur(inverted_img, (21 ,21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

##**Final Sketch**

```python
def dodgeV2(x,y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img=dodgeV2(gray_scale_img , img_smoothing)
cv2_imshow(final_img)
```
